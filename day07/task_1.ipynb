- 입력 중 숫자가 아닌 값이 입력되면 ValueError를 처리
- 숫자 외 값이 입력된 경우 어떤 값이 잘못 입력되었는지 사용자에게 알려주는 문구 출력<br>
  (예: f"잘못된 값: {변수}. 숫자를 입력해주세요.")



````python
save_nums = []
while True:
    try:
       user_input=input('숫자를 입력해주세요 예시 : 1 2 3 4 .. , 종료를 원할시 "done"을 입력')
       if user_input.lower() == 'done':
        break
       nums=user_input.split(" ")
       nums = list(map(int,nums))
       save_nums.extend(nums)
       print(f"저장된 값 : {nums}")
       print(f"전체 저장되 값:{save_nums}")

    except ValueError:
       for num in nums:
          if num.isdigit() == 0:
            print(f"잘못된 값: {num}. 숫자를 입력해 주세요")
    
````

````python
class NegativeNumberError(Exception):
    def __init__(self, message:str)->None:
        self.error_type =self.__class__.__name__
        self.message = message
        super().__init__(message)
    def __str__(self)->str:
        return f"{self.error_type} : {self.message}"
try:
    user_input =int(input("양수를 입력해 주세요"))
    if user_input < 0:
        raise NegativeNumberError("양수를 입력해 주세요")
except NegativeNumberError as e:
    print(e)
    
````

[문제 7]

함수 실행 시간을 측정하는 데코레이터 timer_decorator를 작성하세요. 

임의의 함수가 실행되는 시간을 측정하여 "함수 실행 시간: x.xx초" 형태로 출력하세요.

````python
import time 
def timer_decorator(function):
    def wrapper(*args,**kwargs):
        start = time.time()
        fuc = function(*args,**kwargs)
        end = time.time()
        print(f"실행 시간 : {end - start}")
        return fuc
    return wrapper
@ timer_decorator
def any_func(n:int)->int:
    sum = 0
    for i in range(n):
        sum += i
    return sum
any_func(100)

````

[문제 8]

여러분이 자주 사용하는 수학적 연산<br>
(예: 삼각형 넓이, 원의 넓이, 직육면체의 넓이)을 수행하는 함수들을 포함하는 mymath.py라는 이름의 모듈을 만드세요. 

이 모듈을 사용하여 작성된 함수들을 실제로 임포트하고 사용하는 간단한 프로그램을 작성하세요.

````python
%%writefile mymath.py
import math
def tri_area(height:int,base:int)->int:
    '''
    삼각형은 넓이 구하는 함수
    '''
    return 0.5*height*base
def cir_area(radius:int)->int:
    '''
    원의 넓이를 구하는 함수
    '''
    return math.pi * radius ** 2
def cubo_area(height:int,width:int,length:int)->int:
    '''
    직육면체의 겉넓이를 구하는 함수
    '''
    return (width *length + length * height + width * height) * 2

````

````python
import mymath as mm
triangle =mm.tri_area(10,10)
circle = mm.cir_area(10)
cuboid = mm.cubo_area(10,10,20)
print(f"삼각형:{triangle}, 원:{circle}, 직육면체:{cuboid}")
````

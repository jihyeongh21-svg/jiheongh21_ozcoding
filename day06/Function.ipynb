# 함수

```python
### 함수의 기본 구조
def oz(message, n):
    for i in range(n):
        print(message)
oz("Hello, World!", 5)
```

함수의 기분구조 

def : definition 정의함수 호출  함수

oz :  함수명 

message, n: parameter  매개변수

return : 출력

함수란?

입력(매개변수,parameter)을  받아 특정 작업(logic)실행하여 출력(return,결과)를 받환하는 코드 집합

매개변수(parameter)란?

함수가 특정작업을 하기위해 필요한 입력 값

식별자란? 

함수에서 입력받는 값을 구분하기 위한 별칭

사용이유는?

여러개의 입력값중에 필요한 입력값만 작업의 특정 부분에서 사용하기 위해

함수내에서 어떻게 사용되나?

해다 별칭이 붙은 입력값을 필요한 위치에 별칭을 넣어서 사용함




```python
### 가변매개 변수
def oz2(n,*values):
    for i in range(n):
        print(values)
oz2(3, "Hello", "World", "Python")

```

```python
### 가변매개 변수
def oz2(*values,n): # 가변 매개 변수는 입반 변수뒤에 올수 있음
    for i in range(n):
        print(values)
oz2(3, "Hello", "World", "Python")
```

가변 매개 변수(*args)를 사용하는 이유?

함수에 입력값을 여러개를 동시에 받기 위해서

가변 매개 변수를 사용하기 위한 문법은?

식별자 앞에 '*'를 붙여서 사용한다

가변 매개 변수를 사용하기 위한 제약 조건은?

일반 매개변수를 같이 사용할 경우 일반 매개 변수 뒤에 와야 하면 1번만 사용이 가능

이유?

가변 매개 변수를 호출하여 입력 받을 때 가변 매개 변수가 앞에 있으면 일반 매개 변수와 구분이 되지 않기 때문
 마찬가지의 이유로 2개 이상의 가변 매개 변수도 구분이 되지 않기 때문
 

```python
### 기본 매개 변수 
def oz3(value, n=2):
    for i in range(n):
        print(value)
oz3("Hello, World!")  # n의 값을 지정하지 않으면 기본값 2가 사용됨
```

기본 매개 변수란?

함수를 정의할때 입력값들중 초기에 설정된 입력값

역할은?

함수를 호출해서 사용할 때 기본 매개변수를 입력하지 않아도 설정된 값을 입력 받아 
결과를 출력하게 함

문법은?

일반 매개변수 뒤에 기본 매개변수를 지정하여 사용


제약 조건은 ?

일반 매개 변수뒤에 와야함 이유는 가변 매개변수와 동일 구분의 모호함 때문


```python
### 키워드 매개 변수
def oz4(n=3,*values):
    for i in range(n):
        for value in values:
            print(value)
oz4("hello","world","friend",10)
```

```python
### 키워드 매개 변수
def oz4(*values,n=3):
    for i in range(n):
        for value in values:
            print(value)
oz4("hello","world","friend",n=10)
```

```python
def add_num(a=1,b=2,c=3):
   return a+b+c

print(add_num())
add_num(a=10,b=20,c=30)
```

키워드 매개변수란 

함수를 호출하여 입력값을 넣을 때 특정 실별자를 지정하여 직접 값을 집어넣는 방식

사용이유

가변매개 변수를 앞에 입반 매개 변수를 입력할 떄 둘을 구분짓기 위해 사용

키원드 매개 변수 문법

일반 매개 변수 식별차를 직정하여 해당 식별자 직접 값을 집어 넣는다

가변 키워드 매개 변수
**kwargs
'**'뒤에 딕셔너리 형태로 (식별자:값 , 식별자,:값) 형태로 키워드 매개변수를 입력

```python
### Return 출력 결과
def oz_return():
    print('프린트 시작')
    print("두번째")
    print('세번째')
    return
    print("네번째")
oz_return()
```

```python
def oz_return2():
    return "결과"
result= oz_return2()
print(result)
```

```python
def oz_return3():
    return
result= oz_return3()
print(result)
```

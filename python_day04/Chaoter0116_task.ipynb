__[문제 2: 붕어빵 주문 받기]__

손님이 주문한 붕어빵의 종류와 개수를 입력받는 기능을 구현하세요.

요구사항:
- 손님이 주문할 붕어빵의 종류와 개수를 입력받습니다.
- 주문 내용을 딕셔너리에 저장하세요.
- 주문 내역을 출력해주세요.

```python
orders = dict()
kind = input("붕어빵 종류를 입력하세요:  (종류 :팥붕어빵, 슈크림붕어빵, 초코붕어빵) ")
count = int(input("붕어빵 개수를 입력하세요: "))
orders[kind] = count
print(orders)
```

__[문제 3: 붕어빵 주문 처리]__

손님의 주문을 처리하여 붕어빵 재고를 업데이트하는 기능을 구현하세요.

요구사항:
- 손님의 주문 내용을 기반으로 재고를 업데이트합니다.
- 재고가 부족할 경우, 손님에게 재고 부족을 알리고 재고를 감소시키지 않습니다.
- 재고가 충분할 경우, 주문한 만큼 재고를 감소시키고 판매를 완료합니다.
- 판매가 완료된 경우 판매된 붕어빵 맛과 개수를 출력하세요.

```python
if bread[kind] < count:
     print(f"{kind}의 재고가 부족합니다.")
else:
     bread[kind] -= count
     print(f"{kind} {count}개 주문되었습니다.")
```

__[문제 4: 여러 손님 주문 처리]__

여러 손님들의 주문을 처리할 수 있도록 시스템을 확장하세요.

요구사항:
- 손님이 "종료"를 입력할 때까지 주문을 계속 받습니다.
- 각 주문마다 재고를 업데이트합니다.
- 현재 재고를 화면에 출력합니다.

```python
while True:
    menu = input('1. 주문하기 2.종료 ')
    menu = menu.strip()
    if menu in ('1' ,'주문하기'):
        while True:
            kind = input("붕어빵 종류를 입력하세요:  (종류 :팥붕어빵, 슈크림붕어빵, 초코붕어빵) ")
            kind = kind.strip()
            if kind not in bread.keys():
                print("없는 종류입니다. 다시 입력해주세요.")
                continue
            count = int(input("붕어빵 개수를 입력하세요: "))
            if bread[kind] < count:
                print(f"{kind}의 재고가 부족합니다.")
            else:
                bread[kind] -= count
                orders[kind] = orders.get(kind,0) + count
                print(f"{kind} {count}개 주문되었습니다.")
            break
    elif menu in ('2','종료'):
        print("주문이 종료되었습니다.")
        print("주문 내역:", orders)
        break
    else:
        print("잘못된 입력입니다. 다시 선택해주세요.")    
        continue

```

__[문제 5: 관리자 모드]__

관리자 모드를 통해 붕어빵 재고를 추가할 수 있는 기능을 구현하세요.

요구사항:
- 관리자는 특정 붕어빵의 재고를 추가할 수 있습니다.
- "종료"를 입력하면 관리자 모드를 종료합니다.
- 추가된 재고를 업데이트하고 현재 재고를 화면에 출력합니다.

```python
while True:
    menu = input('1.재고관리 2.종료 ')
    menu = menu.strip()
    if menu in ('1' ,'재고관리'):
        print("현재 재고:", bread)
        while True:
            kind = input("추가할 붕어빵 종류를 입력하세요:  (종류 :팥붕어빵, 슈크림붕어빵, 초코붕어빵) ")
            kind = kind.strip()
            if kind not in bread.keys():
                print("없는 종류입니다. 다시 입력해주세요.")
                continue
            count = int(input("추가할 붕어빵 개수를 입력하세요: "))
            bread[kind] = bread.get(kind,0) + count
            print(f"{kind} {count}개 재고가 추가되었습니다.")
            print("현재 재고:", bread)
            break 
    elif menu in ('2','종료'):
        print("재고 관리가 종료되었습니다.")
        break
```

__[문제 6: 매출 확인]__

    - "초코붕어빵": 1500

- 판매 개수

```python
price = {'팥붕어빵':1000, '슈크림붕어빵':1200, '초코붕어빵':1500}
total = 0
for kind, count in orders.items():
    total += price[kind] * count
print("총 주문 금액은", total, "원입니다.")
```

```python
### 전체 붕어빵 판매프로그램 통합
# 관리자 메뉴 진입 비밀번호
passwored = "admin123"
# 전체 붕어빵 재고 딕셔너리
bread = {"팥붕어빵":10,'슈크림붕어빵':8,'초코붕어빵':5}
# 붕어빵 가격 딕셔너리
price = {'팥붕어빵':1000, '슈크림붕어빵':1200, '초코붕어빵':1500}
# 전체 주문 내역 딕셔너리
orders = dict()
# 전체 매출 변수
total_sales = 0

while True:
    user_type = input('1.고객 2.관리자 3.종료 ')
    user_type = user_type.strip()
    # 고객 메뉴
    if user_type in ('1','고객'):
        while True:
            kind = input("붕어빵 종류를 입력하세요:  (종류 :팥붕어빵, 슈크림붕어빵, 초코붕어빵) ")
            kind = kind.strip()
            if kind not in bread.keys():
                print("없는 종류입니다. 다시 입력해주세요.")
                continue
            count = int(input("붕어빵 개수를 입력하세요: "))
            if bread[kind] < count:
                print(f"{kind}의 재고가 부족합니다.")
            else:
                bread[kind] -= count
                orders[kind] = orders.get(kind,0) + count
                sale_amount = price[kind] * count
                total_sales += sale_amount
                print(f"{kind} {count}개 주문되었습니다. 금액: {sale_amount}원")
            break
    # 관리자 메뉴
    elif user_type in ('2','관리자'):
        pwd = input("관리자 비밀번호를 입력하세요: ")
        pwd = pwd.strip()
        if pwd != passwored:
            print("비밀번호가 틀렸습니다.")
            continue
        elif pwd == passwored :
            while True:
                admin_menu = input('1.재고관리 2.매출확인 3.종료 ')
                admin_menu = admin_menu.strip()
                if admin_menu in ('1' ,'재고관리'):
                    print("현재 재고:", bread)
                    while True:
                        kind = input("추가할 붕어빵 종류를 입력하세요:  (종류 :팥붕어빵, 슈크림붕어빵, 초코붕어빵) ")
                        kind = kind.strip()
                        if kind not in bread.keys():
                            print("없는 종류입니다. 다시 입력해주세요.")
                            continue
                        count = int(input("추가할 붕어빵 개수를 입력하세요: "))
                        bread[kind] = bread.get(kind,0) + count 
                        print(f"{kind} {count}개 재고가 추가되었습니다.")
                        print("현재 재고:", bread)
                        break 
                elif admin_menu in ('2','매출확인'):
                    print("주문 내역:", orders)
                    print("총 매출 금액은", total_sales, "원입니다.")
                elif admin_menu in ('3','종료'):
                    print("관리자 모드가 종료되었습니다.")
                    break
                else:
                    print("잘못된 입력입니다. 다시 선택해주세요.")    
                    continue
    
    elif user_type in ('3','종료'):
        print("프로그램이 종료되었습니다.")
        break
    else:
        print("잘못된 입력입니다. 다시 선택해주세요.")    
        continue
```

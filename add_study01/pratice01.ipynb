"새" RPG 프로그램을 만들려고 합니다.
"새" 의 종류는 "앵무새", "참새", "비둘기", "닭", "러버덕"(오리인형) 이 존재합니다.

각 새는 다른 종류의 "울음소리" 를 가지고 있습니다.

[ 앵무새 ]
울음소리: 까악

[ 참새 ]
울음소리: 짹짹

[ 비둘기 ]
울음소리: 구구

[ 닭 ]
울음소리: 꼬끼오

[ 러버덕 ]
울음소리: 삑삑삑

각 새는 다른 방식으로 "날라다니는 방법" 를 가지고 있습니다.

[ 앵무새 ]
날라다니는 방법: 날개를 힘차게

[ 참새 ]
날라다니는 방법: 날개를 빠르게

[ 비둘기 ]
날라다니는 방법: 날개를 부드럽게

[ 닭 ]
날라다니는 방법: 날개를 퍼덕이며

[ 러버덕 ]
날라다니는 방법: 날지 못함


각 새를 이용해서 "나는 행동"을 하면 "날라다니는 방법" 에 따라 Nm 만큼 나는 행동을 합니다.

새가 날면 "어떤 종류"의 새와 "울음소리"와 어떤 방식으로 "날라다니는 방법"을 선택했는지 몇 "m" 날았는지 출력하는 프로그램을 만드세요.

================예제=======================

[[ 입력 ]]

앵무새


[[ 출력 결과 ]]

앵무새 출발!!!
까악
날게를 힘차게 날었습니다.
결과는 2 m 입니다.


[[ 입력 2 ]]

참새


[[ 출력 결과 2 ]]

참새 출발!!!
짹짹
날게를 빠르게 날았습니다.
결과는 8 m 입니다.


[[ 입력 3 ]]

비둘기




```python
import random
bird ={
    '앵무새':{'소리':"까악",'날기':'날개를 힘차게'},
    '참새':{'소리':"짹짹",'날기':'날개를 빠르게'},
    '비둘기':{'소리':"구구",'날기':'날개를 부드럽게'},
    '닭':{'소리':"꼬끼오",'날기':'날개를 퍼덕이여'},
    '러버덕':{'소리':"삑삑삑",'날기':'날지못함'}

}
for _ in range(6):
    spieces=input('새 종류를 입력해 주세요')
    if spieces not in bird.keys():
        print('등록되지 않은 종입니다')
    else:
        info = bird[spieces]
        print(f"{spieces} 출발")
        print(f"{info['소리']}")
        if spieces !='러버덕':
            print(f"{info['날기']}날았습니다")
            print(f"결과는{random.randrange(1,11)}m입니다")
            
        else:
            print(f"{info['날기']}")
```

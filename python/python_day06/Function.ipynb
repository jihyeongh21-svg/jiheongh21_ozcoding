# 함수

```python
### 함수의 기본 구조
def oz(message, n):
    for i in range(n):
        print(message)
oz("Hello, World!", 5)
```

함수의 기분구조 

def : definition 정의함수 호출  함수

oz :  함수명 

message, n: parameter  매개변수

return : 출력

함수란?

입력(매개변수,parameter)을  받아 특정 작업(logic)실행하여 출력(return,결과)를 받환하는 코드 집합

매개변수(parameter)란?

함수가 특정작업을 하기위해 필요한 입력 값

식별자란? 

함수에서 입력받는 값을 구분하기 위한 별칭

사용이유는?

여러개의 입력값중에 필요한 입력값만 작업의 특정 부분에서 사용하기 위해

함수내에서 어떻게 사용되나?

해다 별칭이 붙은 입력값을 필요한 위치에 별칭을 넣어서 사용함




```python
### 가변매개 변수
def oz2(n,*values):
    for i in range(n):
        print(values)
oz2(3, "Hello", "World", "Python")

```

```python
### 가변매개 변수
def oz2(*values,n): # 가변 매개 변수는 입반 변수뒤에 올수 있음
    for i in range(n):
        print(values)
oz2(3, "Hello", "World", "Python")
```

가변 매개 변수(*args)를 사용하는 이유?

함수에 입력값을 여러개를 동시에 받기 위해서

가변 매개 변수를 사용하기 위한 문법은?

식별자 앞에 '*'를 붙여서 사용한다

가변 매개 변수를 사용하기 위한 제약 조건은?

일반 매개변수를 같이 사용할 경우 일반 매개 변수 뒤에 와야 하면 1번만 사용이 가능

이유?

가변 매개 변수를 호출하여 입력 받을 때 가변 매개 변수가 앞에 있으면 일반 매개 변수와 구분이 되지 않기 때문
 마찬가지의 이유로 2개 이상의 가변 매개 변수도 구분이 되지 않기 때문
 

```python
### 기본 매개 변수 
def oz3(value, n=2):
    for i in range(n):
        print(value)
oz3("Hello, World!")  # n의 값을 지정하지 않으면 기본값 2가 사용됨
```

기본 매개 변수란?

함수를 정의할때 입력값들중 초기에 설정된 입력값

역할은?

함수를 호출해서 사용할 때 기본 매개변수를 입력하지 않아도 설정된 값을 입력 받아 
결과를 출력하게 함

문법은?

일반 매개변수 뒤에 기본 매개변수를 지정하여 사용


제약 조건은 ?

일반 매개 변수뒤에 와야함 이유는 가변 매개변수와 동일 구분의 모호함 때문


```python
### 키워드 매개 변수
def oz4(n=3,*values):
    for i in range(n):
        for value in values:
            print(value)
oz4("hello","world","friend",10)
```

```python
### 키워드 매개 변수
def oz4(*values,n=3):
    for i in range(n):
        for value in values:
            print(value)
oz4("hello","world","friend",n=10)
```

return은?

함수내의 로직의 종료지점 호출지멈을 돌아가며 동시에 함수의 작업의 결과를 반환는 명령어 

```python
### 재귀함수 함수 안에 자신의 함수를 호출하는 함수
def oz_factorial(n):
    if n == 0:
        return 1
    else:
        return n * oz_factorial(n-1)
result = oz_factorial(5)
print(result)
```

재귀함수

재귀함수는 함수안 자신의 함수를 호출하여 사용함으로 반복적인 작업를 번복문을 사용하지 않고
작업이 가능하게 해 코드를 간결하게 하는 장점이 있다

수가 호출될 때마다 컴퓨터의 **'스택(Stack)'**이라는 메모리 공간에 함수의 정보(매개변수, 지역변수 등)가 쌓입니다. 재귀가 너무 깊어지면 이 공간이 꽉 차서 프로그램이 터지게 됩니다.

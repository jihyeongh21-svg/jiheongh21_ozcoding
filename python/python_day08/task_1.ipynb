'Rectangle' 클래스를 작성하세요. 

이 클래스는 높이(height)와 너비(width)를 속성으로 가지며, 생성자(__init__)를 통해 이 두 속성을 초기화합니다. 

'area' 메소드를 추가하여 사각형의 면적을 계산하고 반환하도록 하세요. 높이가 10, 너비가 5인 사각형 객체를 생성하고,<br>

```python
class Rantangle:
    def __init__(self,height:int,width:int)->None:
        self.height = height
        self.width = width
    def area(self)->int:
        return self.height*self.width
    
rt=Rantangle(10,5)
print(rt.area())
        
```

```python
class Employee:
    raise_amount = 1.04

    def __init__(self,name:str,salary:int)->None:
        self.name = name
        self.salary = salary

    @classmethod
    def set_raise_amount(cls,value:float)->None:
        cls.raise_amount = value

    def apply_raise(self)->None:
       self.salary = self.raise_amount*self.salary

em1 =Employee("a",1000)
em2 =Employee('b',1000)
em1.apply_raise()
Employee.set_raise_amount(2.0)
em2.apply_raise()


print(f"{em1.name}인상된 월급{em1.salary}")
print(f"{em2.name}인상된 월급{em2.salary}")
```

[문제 5]

'Account' 클래스를 작성하세요. 

이 클래스는 소유자(owner)와 잔액(balance)을 속성으로 가지며, 잔액은 프라이빗 변수로 선언합니다. 

'deposit'와 'withdraw' 두 개의 메소드를 통해 잔액을 조작할 수 있도록 하세요. 

이 때, 잔액은 음수가 될 수 없도록 체크해야 합니다. 'Account' 객체를 생성하고,<br>
입금 및 출금 작업을 시뮬레이션한 후, 최종 잔액을 출력하세요.

```python
class Account:
    def __init__(self,owner:str,balance:int):
        self.owner = owner
        self.__balance = balance
    def deposit(self,money:int)->None:
        if money > 0:
            self.__balance = self.__balance + money
        else:
            print("정상적인 입금이 아닙니다")
    def withdraw(self,money:int)->None:
        if self.__balance - money > 0 :
            self.__balance = self.__balance - money
        else:
            print("잔액이 부족합니다")

    def get_balance(self):
        return self.__balance
    
 ###############
# 클래스 시물레이션 #
 ###############
account = Account("삼성",10000)
account.deposit(-10)
account.deposit(100)
account.withdraw(200000)
account.withdraw(10000)
print(f"현재 잔액은 {account.get_balance()}원")
         
    

    
```

[문제 6] 

'Vehicle'이라는 기본 클래스를 선언하세요. 

이 클래스는 make (제조사)와 model (모델명) 두 개의 속성을 갖고, 생성자(__init__)를 통해 이 두 속성을 초기화합니다. 'display_info'라는 메소드를 추가하여, "이 차량은 [make]에서 제조한 [model]입니다."라는 메시지를 반환하도록 하세요.

'Truck'과 'Car' 두 개의 하위 클래스를 생성하고, 각 클래스는 'Vehicle' 클래스로부터 상속받습니다. 

'Truck' 클래스에는 추가로 'payload' (적재량) 속성을, 'Car' 클래스에는 'seats' (좌석 수) 속성을 추가합니다. 각각의 하위 클래스에서 'display_info' 메소드는 기본 클래스의 메소드를 그대로 사용합니다.

'Truck'과 'Car' 객체 각각을 생성하고, 'display_info' 메소드를 호출하여 정보를 출력하세요. 

또한, 각 객체의 추가 속성(적재량, 좌석 수)에 대한 정보도 출력하세요.

이 문제는 클래스의 상속 개념을 이해하고, 기본 클래스의 속성과 메소드를 하위 클래스가 어떻게 활용할 수 있는지 보여줍니다. 

오버라이드는 필요하지 않으며, 각 하위 클래스의 고유한 속성을 추가하는 방법에 초점을 맞춥니다.

```python
class Vehicle:
    def __init__(self,make,model):
        self.make = make
        self.model = model
    def display_info(self):
        return f"제조사: [{self.make}], 모델:[{self.model}]"
class Truck(Vehicle):
    def __init__(self, make, model,carry):
        super().__init__(make, model)
        self.carry = carry
    def display_truck(self):
        return f"적재중량은 {self.carry} kg 입니다"

class Car(Vehicle):
    def __init__(self, make, model,seats):
        super().__init__(make, model)
        self.seats= seats
    def display_car(self):
        return f"좌석수는 {self.seats}좌석입니다"
truck = Truck("기아","봉고3",20000)
car = Car("현대","소나타",4)
print(truck.display_info())
print(truck.display_truck())
print(car.display_info())
print(car.display_car())
    
```

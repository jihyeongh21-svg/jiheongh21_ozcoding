```python
import requests

url = "https://httpbin.org/get"
r = requests.get(url, timeout=10)  # timeout은 습관처럼
r.raise_for_status()               # 4xx/5xx면 예외로 즉시 실패 처리

print(r.status_code)
print(r.text[:200])
```

```python
url = "https://httpbin.org/get"
params = {"query": "손흥민", "page": 1}

r = requests.get(url, params=params, timeout=10)
r.raise_for_status()
print(r.url)  # 최종 URL 확인(디버깅에 매우 유용)
```

```python
import requests
from bs4 import BeautifulSoup
# 1. URL 설정 및 쿼리 스트링
# keyword = input("검색어를 입력하세요: ")
url = "https://search.naver.com/search.naver"
params = {
"where":"nexearch",
"sm":"top_hty",
"fbm":"0",
"ie":"utf8",
"query":"손흥민",
"ackey":"3c1bpylg"
}
res = requests.get(url,params=params)
res.raise_for_status()
# print(res.text)
html = BeautifulSoup(res.content,"lxml")
target = html.select_one("#main_pack > section.sc_new.cs_common_module.case_normal._au_people_content_wrap.color_14._cs_sports_people")
real_target = target.select_one("div.cm_info_box")
all_dts = real_target.select("dt")
all_dds = real_target.select("dd")
element_len = len(all_dts)
# for x in all_dds:
#     print(x.get_text())
result_set = dict()
for idx in range(element_len):
    key = all_dts[idx].get_text().strip()
    value = all_dds[idx].get_text().strip()
    result_set[key]=value
print(result_set)
```
